```python
from glob import glob
import pandas as pd

import altair as alt
```

## Read Data

```python
result_filepath = glob("*/every_eurovision_result_ever.csv")
```

```python
results = pd.read_csv(result_filepath[0])
```

```python
results.info()
```

```python
results[results["Grand Final Place"].isna()] # 1956 was the first year of competition and the votes were lost. 2020 was cancelled due to COVID
```

```python
# Convert the 'Language' column to lists
results['Language'] = results['Language'].apply(lambda x: x.split(', '))
```

```python
results = results.explode(column="Language")
```

```python
results['Language'] = results['Language'].str.replace(r'\[\w+\]', '', regex=True)
```

```python
results.sort_values(by="Song").head()
```

```python
results.groupby(by="Language").nunique().sort_values(by="Song", ascending=False).head(60)
```

```python
filtered_results = results.groupby('Language').filter(lambda x: len(x) > 6)
```

```python
filtered_results.head()
```

```python
alt.Chart(filtered_results).mark_bar().encode(
    x="Year",
    y="Song",
    color="Language"
)
```
